Compute Accuracy

```python
from seqeval.metrics import accuracy_score
from seqeval.metrics import classification_report
from seqeval.metrics import f1_score
test_pre_list = []
test_list = []

temp_list = []
file_test_pre = open("output/2.2train_test_predictions.txt", "r")
for line in file_test_pre:
    if line == "\n":
        test_pre_list.append(temp_list)
        temp_list = []
        continue
    temp_list.append(line.split(" ")[1].split("\n")[0].replace("X", "O"))
file_test_pre.close()

file_test = open("data/test.txt", "r")
for line in file_test:
    if line == "\n":
        test_list.append(temp_list)
        temp_list = []
        continue
    try:
        temp_list.append(line.split(" ")[1].split("\n")[0])
    except:
        temp_list.append("O")
file_test.close()

print(classification_report(test_list, test_pre_list))
```
